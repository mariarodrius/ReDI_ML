```python
#this code works
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
mydata = np.loadtxt(fname="data0.txt")
mydata
```

```python
plt.scatter(mydata[:,0],mydata[:,1], c=mydata[:,2])  #c for color in 3rd row (position 2)
```

```python
class LinearClassifier:
    
    def __init__(self):
        print("Initializing")
        
    def train(self, Xtrain):
        print("training")
        self.Xtrain = Xtrain
        Xvalues = np.linspace(0,18,20)
        plt.scatter(mydata[:,0], mydata[:,1], c=mydata[:,2])
        # select only data 0
        X = self.Xtrain[np.where(self.Xtrain[:,2]==0),0][0]
        Y = self.Xtrain[np.where(self.Xtrain[:,2]==0),1][0]
        print(Y)
        A = np.vstack([X, np.ones(len(X))]).T 
        solu = np.linalg.lstsq(A,Y,rcond=None)    #############
        m,b=solu[0]
        yhat = m *Xvalues +b
        plt.plot(Xvalues,yhat,c="black")
        
        #select only data 1, yellow dots
        X = self.Xtrain[np.where(self.Xtrain[:,2]==1),0][0]
        Y = self.Xtrain[np.where(self.Xtrain[:,2]==1),1][0]
        print(Y)
        A = np.vstack([X, np.ones(len(X))]).T 
        solu = np.linalg.lstsq(A,Y,rcond=None)
        my,by=solu[0]
        yhat = my *Xvalues +by
        plt.plot(Xvalues,yhat,c="yellow")
        
        m = 0.1
        b = 12
             
        #take the mean value of the two m and b 
        self.mm = (m+my)/2                   ###### need the self. part to define it within the class - classifier
        self.bb = (b+by)/2     
    
        
    def predict(self, Xtest):
        print("Linear Classifier Inference")
        plt.scatter(Xtest[0],Xtest[1],s=100,c=["red"])
        if self.mm * Xtest[0] + self.bb > Xtest[1]:
            print("Test data is purple")
        else:
            print("Test data is yellow")
```

```python
classificator = LinearClassifier()
classificator.train(mydata)
classificator.predict([8.0, 10.0])   ### 8.0 in x axis and 10.0 in y axis. check red dot position in plot below
```
