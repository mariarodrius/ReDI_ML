```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
def func_sigmoid(x): #logistic function
    return 1.0 / (1.0 + np.exp(-x))
```

```python
def func_tanh(x): #hyperbolic tangent
    return (np.exp(x)-np.exp(-x)/np.exp(x)+np.exp(-x))
```

```python
def func_relu(x): #hyperbolic tangent
    return np.where(x>0,x,0)
```

```python
def func_leakyrelu(x, a=0.05): #to solve gradients, choose the value of a to show the gradual step
    return np.where(x>0,x,x*a)
```

```python
x = np.arange(-4,4,0.1)
plt.plot(x,func_sigmoid(x),linewidth=5.0)
plt.title("logisticv0")
```

```python
plt.subplots_adjust(wspace=0.4, hspace=0.4)
x = np.arange(-4,4,0.1)
plt.subplot(231)
plt.plot(x,func_sigmoid(x),linewidth=5.0)
plt.title("logistic")
plt.subplot(232)
plt.plot(x,func_tanh(x),linewidth=5.0)
plt.title("tanh")
plt.subplot(234)
plt.plot(x,func_relu(x),linewidth=5.0)
plt.title("relu")
plt.subplot(235)
plt.plot(x,func_leakyrelu(x),linewidth=5.0)
plt.title("leakyrelu")
```
